```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Raw Sales Data.csv")
df
```

```python
df.info()
```

```python
df['Qty'] = df['Qty'].str.split(',').str.join("").astype(int)  # str.join("") -- delimiter
```

```python
df['Sales'] = df['Sales'].str.split(',').str.join("").astype(int)
```

```python
df['Month'] = pd.to_datetime(df['Month'])
```

```python
df.info()
```

```python
df.head()
```

```python
# the "strftime('%B')" method is used to format the datetime object to return the full month name. The strftime function stands for "string format time". 

# %Y - Year with century as a decimal number (e.g., 2024)
# %B - Full month name (e.g., January, February, etc.)
# %b - Abbreviated month name (e.g., Jan, Feb, etc.)
# %A - Full weekday name (e.g., Monday, Tuesday, etc.)
# %a - Abbreviated weekday name (e.g., Mon, Tue, etc.)
```

```python
df['Month_Name'] = df['Month'].dt.strftime('%B')
```

EXERCISE

```python
# Question-1. Monthwise number of quantities sold

Q_1 = df.groupby(['Month','Month_Name'])['Qty'].sum().reset_index()
Q_1
```

```python
# Question-2. Category wise Qty sold.

Q_2 = df.groupby(['Category'])['Qty'].sum().reset_index()
Q_2
```

```python
#  Question-3. Sub-channel wise total sales.

Q_3 = df.groupby(['Sub-Channels'])['Sales'].sum().reset_index()
Q_3
```

```python
# Question-4. Highest sale in which month.

df_mon_sales = df.groupby(['Month','Month_Name'])['Sales'].sum()
df_max_sales = df_mon_sales.max()
df_max_month = df_mon_sales.idxmax()

df_result = pd.DataFrame({'Month' : [df_max_month],
                         'Sales': [df_max_sales]})
print(df_result)
```

```python
import re

product_name = "GET SLIM JUICE 500 ML"
cleaned_product_name = re.sub(r'\d+ ?[A-Za-z]+$', '', product_name).strip()
print(cleaned_product_name)

# re.sub(pattern, replacement, string)

#'\d+ ?[A-Za-z]+$'

# \d+ - \d matches any digit (0-9),This part matches the numeric quantity in the product name (e.g., 1, 500, 250).

# ? - This part matches an optional space that may follow the numeric quantity.

# [A-Za-z]+ - [] defines a character class.A-Za-z matches any uppercase (A-Z) or lowercase (a-z) letter. This part matches the packaging unit (e.g., L, ML).

# $ - This asserts the position at the end of the string. Ensures that the pattern matches only if it's at the end of the product name.
```

```python
# Question-5. Highest selling product (based on sales) (packaging quantity doesn't matter. Ex. consider "GET SLIM JUICE 1 L" and "GET SLIM JUICE 500 ML" as one product)

df['Cleaned_product'] = df['Product Name'].apply(lambda x: re.sub(r'\d+ ?[A-Za-z]+$','',x).strip())

df_cleaned = df.groupby(['Cleaned_product'])['Sales'].sum().reset_index()

max_sale_product = df_cleaned.loc[df_cleaned['Sales'].idxmax()]

print(max_sale_product)
```
